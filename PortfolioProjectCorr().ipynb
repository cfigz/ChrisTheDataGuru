```python
#import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) #Adjust the configuration of the plots we will create

# Read in the data

df = pd.read_csv('/Users/Shared/Downloads/Data Analysis Python/movies.csv')
```

```python
# Examine the data
df.head()
```

```python
# Look for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# Data types of columns

df.dtypes
```

```python

df['budget'] = df['budget'].astype("Int64")
df['gross'] = df['gross'].astype("Int64")
df = df.replace(np.nan, 0)
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
#Drop any duplicates
df['company'] = pd.CategoricalIndex(df['company'], ordered=True)
df['company'].drop_duplicates().sort_values().head()
```

```python
df.head()
```

```python
# Budget high Correlation
# Company high Correlation
```

```python
# Build scatter plot w/ budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget For Film')
plt.show()
```
